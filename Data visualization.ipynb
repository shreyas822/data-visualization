```python
grp=df.groupby('age')
x=grp["balance"].agg(np.sum)
y=grp["duration"].agg(np.mean)
z=grp["campaign"].agg(np.mean)
a=grp["pdays"].agg(np.mean)
b=grp["day"].agg(np.mean)
c=grp["previous"].agg(np.mean)
print(x)
print(y)
print(z)
print(a)
print(b)
print(c)
```

```python
plt.figure(figsize=(12,5))
plt.plot(x, 'ro')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.plot(x,'ro', color='b')
plt.xticks(rotation=90)
plt.title('Agewise bank balance')
plt.xlabel('age-->')
plt.ylabel('balance-->')
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.plot(y,'r--', color='y')
plt.xticks(rotation=90)
plt.title('Agewise bank balance')
plt.xlabel('age-->')
plt.ylabel('duration-->')
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.plot(z,'r--', color='g')
plt.xticks(rotation=90)
plt.title('Agewise bank balance')
plt.xlabel('age-->')
plt.ylabel('campain-->')
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.plot(a,'ro', color='b')
plt.xticks(rotation=90)
plt.title('Agewise bank balance')
plt.xlabel('age-->')
plt.ylabel('pdays-->')
plt.show()
```
